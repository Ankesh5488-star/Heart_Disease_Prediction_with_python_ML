Import the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

data collection and processing

```python
#loading the data to a pandas data frame
heart_data=pd.read_csv('/content/heart_disease_data.csv')
```

splitting the features and target

```python
X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into traning data and testing data

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,x_train.shape,x_test.shape)
```

model traning


```python
#Logistic Regression
model=LogisticRegression()
```

```python
#traning with LogisticRegression model with traning data
model.fit(x_train,y_train)
```

model Evalution

Accuracy Score

```python
#accuracy of the training data
x_train_prediction=model.predict(x_train)
traning_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy of the traning data:',traning_data_accuracy)
```

```python
#accuracy score of the test data
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Acuuracy score for the test data:',testing_data_accuracy)
```

building A predictive System

```python
input_data=(60,1,0,130,253,0,1,144,1,1.4,2,1,3)
#change input data in a numpy array
input_data_as_numpy_array=np.asarray(input_data)

#reshape the numpy array as we are predictive for only one instant
input_data_reshape=input_data_as_numpy_array.reshape(1,-1)

prediction=model.predict(input_data_reshape)
if(prediction[0]==0):
  print("The person does not have any heart disease")

else:
  print("The person has heart disease")

```
